From https://www.geeksforgeeks.org/python-ocr-on-all-the-images-present-in-a-folder-simultaneously/

```python
from PIL import Image
import pytesseract as pt
import os
      
def main():
    # path for the folder for getting the raw images
    path ="/home/juris/Documents/ariadne/models/data/photos/banners"
  
    # link to the file in which output needs to be kept
    fullTempPath ="/home/juris/Documents/ariadne/models/calliope/cpbannertext.txt"
  
    # iterating the images inside the folder
    for imageName in os.listdir(path):
        inputPath = os.path.join(path, imageName)
        img = Image.open(inputPath)
  
        # applying ocr using pytesseract for python
        text = pt.image_to_string(img, lang ="eng")
        text = text.replace("\n", " ") # added to remove line breaks
  
        # saving the  text for appending it to the output.txt file
        # a + parameter used for creating the file if not present
        # and if present then append the text content
        file1 = open(fullTempPath, "a+")
  
        # providing the name of the image
        file1.write(imageName+"\n")
  
        # providing the content in the image
        file1.write(text+"\n")
        file1.close() 
  
    # for printing the output file
    file2 = open(fullTempPath, 'r')
    print(file2.read())
    file2.close()        
  
  
if __name__ == '__main__':
    main()
```

Data processing
- Removing double spaces
- Changing | to I

```python
# Read
with open('/home/juris/Documents/ariadne/models/calliope/cpbannertext.txt', 'r') as file :
  filedata = file.read()

# Replace
filedata = filedata.replace('  ', ' ')
filedata = filedata.replace('|', 'I')

# Write to same file
with open('/home/juris/Documents/ariadne/models/calliope/cpbannertext.txt', 'w') as file:
  file.write(filedata)
```
